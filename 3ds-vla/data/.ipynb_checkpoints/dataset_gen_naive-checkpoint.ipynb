```python
base_dataset_origin = "/home/jiyao/mingxu/LLaMA-Adapter-v3-Demo-release_336_large_Chinese/data/traincat_1113"

output_dataset = "/home/jiyao/xcy/LLaMA-Adapter-v3-Demo-release_336_large_Chinese_ori/data/dataset/naive_pose2pose"
```

```python
import os
import json
import re
import numpy as np

for filename in os.listdir(base_dataset_origin):
    if filename.endswith('.json'):  #  JSON 
        file_path = os.path.join(base_dataset_origin, filename)  # 
        with open(file_path, 'r') as file:  # 
            print(filename)
            data = json.load(file)  #  JSON 

            new_data = {}
            new_data["image"] = data["image"]

            random_list =  np.random.random(8) 
            # mask01
            mask = np.random.randint(0, 2, size=8)
            if mask.sum() == 0:
                mask[np.random.randint(0,5)] = 1
                print(mask)
            mask[6] = mask[7] = mask[5] = 0

            gt_string = data["output"]
            numbers = re.findall(r'-?\d+\.\d+e-?\d+|-?\d+\.\d+|\d+', gt_string)
            gt_list = [float(i) for i in numbers]
            print(gt_list)
            

            # maskrandom_listanother_list
            fail_list = gt_list + random_list * mask

            instruction = f'''I fail to manipulate this object.Correct the contact point and orientation of using the object.Currenct contact point is at ({int(fail_list[0])}, {int(fail_list[1])}),the gripper direction is [{fail_list[2]}, {fail_list[3]}, {fail_list[4]}], the gripper forward direction is [{gt_list[5]}, {gt_list[6]}, {gt_list[7]}].'''
            new_data["conversations"] = data["conversations"]
            new_data["conversations"][0]["prompt"] = instruction
            new_data["instruction"] = instruction
            new_data["input"] = data["input"]
            new_data["output"] = data["output"]

            with open(os.path.join(output_dataset, filename), 'w') as f:
                json.dump(new_data, f)



```
